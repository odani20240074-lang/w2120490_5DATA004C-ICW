```python
# Drop rows with missing values
df = df.dropna()
print(df.isnull().sum().sum())  
```

```python
year_cols_all = [c for c in df.columns if c.isdigit()]
df[year_cols_all].describe().loc[['min', 'max', 'mean']]
```

```python
# Keep only required columns
year_cols = [col for col in df.columns if str(col).isdigit()]

keep_cols = ["REF_AREA", "REF_AREA_LABEL", "URBANISATION", "URBANISATION_LABEL"] + year_cols
df = df[keep_cols].copy()
df.head()
```

```python
#Renaming
df = df.rename(columns={
    "REF_AREA":           "country_code",
    "REF_AREA_LABEL":     "country",
    "URBANISATION":       "area_type_code",
    "URBANISATION_LABEL": "area_type",
})
df.head()
```

```python
df_cleaned = df.melt(
    id_vars=['country_code', 'country', 'area_type_code', 'area_type'],
    var_name='year',
    value_name='value'
)

df_cleaned['year'] = df_cleaned['year'].astype(int)

df_cleaned = df_cleaned.sort_values(['country', 'area_type', 'year']).reset_index(drop=True)

df_cleaned.head(20)
```

```python
# Save cleaned data
df_cleaned.to_csv('sanitation_clean.csv', index=False)
print('Saved')
```
